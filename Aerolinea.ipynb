# Visualizacion N°1

Gráfico de barras que muestra la cantidad de vuelos por aerolínea.


*   El gráfico de barras: Es una visualización que muestra las aerolíneas en el eje x y la cantidad de vuelos en el eje y. Cada barra representa una aerolínea y su altura indica la cantidad de vuelos que ha realizado dicha aerolínea.

*   Importancia del gráfico: El gráfico de barras es útil para comparar visualmente la cantidad de vuelos entre diferentes aerolíneas. Nos permite identificar rápidamente cuáles aerolíneas tienen más vuelos y cuáles tienen menos. Esta información puede ser relevante para realizar análisis, tomar decisiones comerciales o entender la distribución de vuelos en un conjunto de datos.

*  Datos utilizados: El gráfico utiliza la columna 'Airline' del DataFrame df para obtener la información sobre las aerolíneas. La columna 'Airline' debe contener los nombres de las aerolíneas correspondientes a cada vuelo registrado en el conjunto de datos.

En resumen, el gráfico de barras muestra la cantidad de vuelos por aerolínea y permite comparar visualmente la distribución de vuelos entre las diferentes aerolíneas utilizadas en el conjunto de datos.

```python
# Obtener el recuento de vuelos por aerolínea y ordenar en orden descendente
flight_counts = df['Airline'].value_counts().sort_values(ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  # Tamaño del gráfico (opcional)
sns.countplot(x='Airline', data=df, order=flight_counts.index)  # Ordenar en orden descendente
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x (opcional)
plt.xlabel('Aerolínea')
plt.ylabel('Cantidad de vuelos')
plt.title('Cantidad de vuelos por aerolínea')

# Agregar etiquetas de porcentaje
total = len(df['Airline'])
for i, count in enumerate(flight_counts):
    percentage = (count / total) * 100
    plt.text(i, count + 5, f'{percentage:.1f}%', ha='center')


# Mostrar el gráfico
plt.show()
```

# Visualizacion N°2

Gráfico de líneas que muestra la evolución del precio promedio de los billetes de avión a lo largo del tiempo. Aquí está la explicación del código:


*  **df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])**: Esta línea convierte la columna "Date_of_Journey" en el tipo de dato datetime, lo cual es útil para trabajar con fechas en Python.

*  **df_grouped = df.groupby('Date_of_Journey')['Price'].mean().reset_index()**: Aquí, el DataFrame original se agrupa por la columna "Date_of_Journey" y se calcula el promedio de los precios correspondientes a cada fecha. El resultado se almacena en el DataFrame df_grouped.

*  **df_grouped.sort_values('Date_of_Journey', inplace=True)**: Esta línea ordena el DataFrame df_grouped en función de la columna "Date_of_Journey" para asegurar que las fechas se muestren en orden cronológico.

Se crea un gráfico de líneas utilizando sns.lineplot(), donde el eje x representa las fechas de viaje y el eje y representa los precios promedio de los billetes.

*  Este gráfico es útil para visualizar la tendencia general de los precios promedio de los billetes de avión a lo largo del tiempo. Puede ayudar a identificar patrones estacionales, fluctuaciones y cambios a lo largo de diferentes períodos. Esto puede ser valioso para el análisis y la toma de decisiones relacionadas con la industria de la aviación, como la planificación de tarifas o la identificación de períodos de alta demanda y precios más altos.

```python
# Convertir la columna "Date_of_Journey" al tipo de datos datetime
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])

# Agrupar por fecha y calcular el promedio de los precios
df_grouped = df.groupby('Date_of_Journey')['Price'].mean().reset_index()

# Ordenar el DataFrame por la columna "Date_of_Journey"
df_grouped.sort_values('Date_of_Journey', inplace=True)

# Crear el gráfico de líneas con los valores agregados
plt.figure(figsize=(14, 4))  # Tamaño del gráfico (opcional)
sns.lineplot(x='Date_of_Journey', y='Price', data=df_grouped)
plt.xlabel('Fecha de Viaje')
plt.ylabel('Precio promedio del Billete')
plt.title('Evolución del precio promedio de los billetes de avión')

# Mostrar el gráfico
plt.show()
```

# Visualizacion N°3

El gráfico presentado es un gráfico de barras que muestra la frecuencia de palabras clave encontradas en la columna "Additional_Info" de los datos. Cada barra representa una palabra clave y su altura indica la frecuencia con la que aparece en los datos. El gráfico está ordenado de mayor a menor frecuencia y utiliza etiquetas de porcentaje para mostrar el porcentaje de frecuencia de cada palabra clave.


Este gráfico es importante para analizar las palabras clave más comunes en los datos y su relevancia en la columna "Additional_Info". Permite identificar las palabras clave que aparecen con mayor frecuencia y comprender su impacto en el conjunto de datos. Con esta información, se pueden extraer patrones o tendencias relacionadas con las palabras clave y utilizarlas para mejorar el análisis y la toma de decisiones en el contexto del proyecto.

```python
from collections import Counter
import re

# Extraer todas las opciones dentro de Additional_Info
all_options = set([option for sublist in df['Additional_Info'].str.split('|') for option in sublist])

# Extraer las palabras clave basadas en las opciones
keywords = [option for option in all_options if re.search(r'\b\w+\b', option)]

# Crear una nueva columna para las palabras clave
df['Keywords'] = df['Additional_Info'].apply(lambda x: [word for word in keywords if word in x])

# Contar la frecuencia de las palabras clave
keyword_counts = Counter([keyword for sublist in df['Keywords'] for keyword in sublist])

# Crear un DataFrame con las palabras clave y sus frecuencias
df_keywords = pd.DataFrame(keyword_counts.items(), columns=['Keyword', 'Frequency'])

# Ordenar el DataFrame por la frecuencia de mayor a menor
df_keywords.sort_values('Frequency', ascending=False, inplace=True)

# Calcular el porcentaje de frecuencia
total_keywords = df_keywords['Frequency'].sum()
df_keywords['Percentage'] = (df_keywords['Frequency'] / total_keywords) * 100

# Crear el gráfico de barras con etiquetas de porcentaje
plt.figure(figsize=(10, 6))
sns.barplot(x='Keyword', y='Frequency', data=df_keywords, palette='Set3')
plt.xlabel('Palabra clave')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de palabras clave en Additional_Info (ordenado por frecuencia)')

# Agregar etiquetas de porcentaje en las barras
for i, row in df_keywords.iterrows():
    plt.text(i, row['Frequency'], f"{row['Percentage']:.1f}%", ha='center', va='bottom')

# Rotar las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.show()
```

Al conocer los puntos de origen y destino más comunes, las aerolíneas pueden tomar decisiones estratégicas, como aumentar la frecuencia de vuelos en esas rutas populares o ajustar su capacidad en función de la demanda. Por otro lado, para los viajeros, el gráfico les brinda una visión general de los destinos más populares, lo que les permite planificar sus viajes y considerar las opciones de vuelo desde esos lugares.

Además, el gráfico nos muestra la proporción de vuelos para cada ubicación, representada en forma de porcentaje. En resumen, el análisis de este gráfico nos permite comprender mejor los patrones de tráfico aéreo y puede ser utilizado tanto por las aerolíneas como por los viajeros para tomar decisiones informadas relacionadas con los vuelos y los destinos.

```python
# Contar la frecuencia de cada punto de origen y destino
source_counts = df['Source'].value_counts()
destination_counts = df['Destination'].value_counts()

# Crear una lista de etiquetas para el gráfico
labels_source = source_counts.index.tolist()
labels_destination = destination_counts.index.tolist()

# Crear una lista de valores para el gráfico
values_source = source_counts.values.tolist()
values_destination = destination_counts.values.tolist()

# Crear la figura y los subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Graficar el primer gráfico de proporción de vuelos por punto de origen
ax1.pie(values_source, labels=labels_source, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')
ax1.set_title('Proporción de vuelos por punto de origen')

# Graficar el segundo gráfico de proporción de vuelos por punto de destino
ax2.pie(values_destination, labels=labels_destination, autopct='%1.1f%%', startangle=90)
ax2.axis('equal')
ax2.set_title('Proporción de vuelos por punto de destino')

# Ajustar el espaciado horizontal entre los subplots
plt.subplots_adjust(wspace=0.3)

# Mostrar los gráficos
plt.show()
```

# Visualización N°6

El gráfico presentado es un heatmap que representa la frecuencia de vuelos en diferentes horas del día y días de la semana. Cada celda del heatmap muestra el número de vuelos que se realizaron en una hora específica y un día de la semana específico. Los colores en el heatmap representan la intensidad de la frecuencia, donde los tonos más oscuros indican una mayor frecuencia de vuelos y los tonos más claros indican una menor frecuencia.

Este tipo de gráfico es útil para identificar patrones en la distribución de vuelos a lo largo del día y a lo largo de la semana. Puede ayudar a identificar las horas y los días de la semana más ocupados o menos ocupados en términos de vuelos, lo que puede ser relevante para la programación de vuelos, la gestión de la capacidad y la planificación estratégica de las aerolíneas. La visualización de los datos en forma de heatmap facilita la identificación de los momentos de mayor demanda y los momentos de menor actividad.

```python
# Extraer la hora y el día de la semana a partir de la columna "Dep_Time"
df['Dep_Hour'] = pd.to_datetime(df['Dep_Time']).dt.hour
df['Day_of_Week'] = pd.to_datetime(df['Date_of_Journey']).dt.day_name()

# Crear una tabla de contingencia para contar la frecuencia de vuelos en cada hora y día
heatmap_data = df.pivot_table(index='Dep_Hour', columns='Day_of_Week', aggfunc='size', fill_value=0)

# Crear el heatmap
plt.figure(figsize=(12, 8))  # Tamaño del heatmap (opcional)
sns.heatmap(heatmap_data,linecolor = "white",linewidths = 0.05, cmap='Blues', annot=True, fmt='d', cbar=True)

# Personalizar el aspecto del heatmap
plt.title('Frecuencia de vuelos por hora y día de la semana')
plt.xlabel('Día de la semana')
plt.ylabel('Hora del día')

# Mostrar el heatmap
plt.show()
```
